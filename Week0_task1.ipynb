# Summary Statistics 

```python
# calculate mean, median mode, standard deviation 
def calculate_summary_statistics(df):
    # Filter numeric columns (excluding object datatype columns)
    numeric_cols = df.select_dtypes(include=['float64', 'int64'])
    
    # Calculate statistics
    summary_statistics = numeric_cols.describe()
    
    return summary_statistics

# Define your datasets
data_benin = pd.read_csv('benin-malanville.csv')
data_sierraleone = pd.read_csv('sierraleone-bumbuna.csv')
data_togo = pd.read_csv('togo-dapaong_qc.csv')

# Calculate summary statistics for each dataset
summary_benin = calculate_summary_statistics(data_benin)
summary_sierraleone = calculate_summary_statistics(data_sierraleone)
summary_togo = calculate_summary_statistics(data_togo)

# Print summary statistics for each dataset
print("Summary statistics for Benin dataset:")
print(summary_benin)

print("\nSummary statistics for Sierra Leone dataset:")
print(summary_sierraleone)

print("\nSummary statistics for Togo dataset:")
print(summary_togo)
```

# Data Quality Check

```python
import pandas as pd

def check_data_quality(datasets):
    results = []
    
    for i, df in enumerate(datasets, start=1):
        # Create a dictionary to store results for each dataset
        dataset_result = {'Dataset': f'Dataset {i}'}
        
        # Check for missing values
        missing_values = df.isnull().sum()
        dataset_result['Missing Values'] = missing_values.sum()  # Total missing values in the dataset
        
        # Check for negative values in specific columns (GHI, DNI, DHI)
        negative_values_cols = ['GHI', 'DNI', 'DHI']
        negative_values = df[negative_values_cols].lt(0).sum().sum()  # Total negative values in specific columns
        dataset_result['Negative Values'] = negative_values
        
        # Check for outliers (using a simple approach)
        # You can define your own logic for identifying outliers
        outliers_cols = ['GHI', 'DNI', 'DHI']
        q1 = df[outliers_cols].quantile(0.25)
        q3 = df[outliers_cols].quantile(0.75)
        iqr = q3 - q1
        outliers = ((df[outliers_cols] < (q1 - 1.5 * iqr)) | (df[outliers_cols] > (q3 + 1.5 * iqr))).sum().sum()  # Total outliers
        dataset_result['Outliers'] = outliers
        
        # Append the results for the dataset to the list
        results.append(dataset_result)
    
    # Convert the list of dictionaries to a DataFrame
    results_df = pd.DataFrame(results)
    return results_df

# Load your datasets as DataFrame objects
df1 = pd.read_csv('benin-malanville.csv')
df2 = pd.read_csv('sierraleone-bumbuna.csv')
df3 = pd.read_csv('togo-dapaong_qc.csv')

# Pass the datasets to the function
results = check_data_quality([df1, df2, df3])
print(results.reset_index())
```

```python
# check for missing values, outliers, and othre annomalies in the dataset
def check_data_quality(datasets):
    results = []
    
    for i, df in enumerate(datasets, start=1):
        # Create a dictionary to store results for each dataset
        dataset_result = {'Dataset': f'Dataset {i}'}
        
        # Check for missing values
        missing_values = df.isnull().sum()
        dataset_result['Missing Values'] = missing_values.sum()  # Total missing values in the dataset
        
        # Identify columns with missing values
        columns_with_missing_values = missing_values[missing_values > 0].index.tolist()
        dataset_result['Columns with Missing Values'] = columns_with_missing_values
        
        # Check for negative values in specific columns (GHI, DNI, DHI)
        negative_values_cols = ['GHI', 'DNI', 'DHI']
        negative_values = df[negative_values_cols].lt(0).sum().sum()  # Total negative values in specific columns
        dataset_result['Negative Values'] = negative_values
        
        # Check for outliers (using a simple approach)
        # You can define your own logic for identifying outliers
        outliers_cols = ['GHI', 'DNI', 'DHI']
        q1 = df[outliers_cols].quantile(0.25)
        q3 = df[outliers_cols].quantile(0.75)
        iqr = q3 - q1
        outliers = ((df[outliers_cols] < (q1 - 1.5 * iqr)) | (df[outliers_cols] > (q3 + 1.5 * iqr))).sum().sum()  # Total outliers
        dataset_result['Outliers'] = outliers
        
        # Append the results for the dataset to the list
        results.append(dataset_result)
    
    # Convert the list of dictionaries to a DataFrame
    results_df = pd.DataFrame(results)
    return results_df

# Load your datasets as DataFrame objects
df1 = pd.read_csv('benin-malanville.csv')
df2 = pd.read_csv('sierraleone-bumbuna.csv')
df3 = pd.read_csv('togo-dapaong_qc.csv')

# Pass the datasets to the function
results = check_data_quality([df1, df2, df3])
print(results.reset_index())
```
